```python
from simulations import sdn_ecckn, iot_broker, dos_attack, anti_dos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
status, energies = sdn_ecckn.run(6000)
```

```python
df_energy = pd.DataFrame(energies)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df_energy)
plt.title("CONSUMO DE ENERGIA - SDN-ECCKN ORIGINAL", size=15)
plt.xlabel("Ciclos")
plt.xlim(xmin=0.0)
plt.ylabel("Energia dos Sensores [Volts]")
plt.ylim(ymin=2.05)
plt.show()
```
